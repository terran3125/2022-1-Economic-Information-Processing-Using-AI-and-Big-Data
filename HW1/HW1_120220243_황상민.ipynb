```python
"""
작성자 성명: 황상민 120220243

이 Jupyter notebook이 하는 내용 설명

저는 통계청에서 제공한 2020년 시도별 1인당 지역내총생산, 지역총소득, 개인소득 데이터와
(출처: https://kosis.kr/statHtml/statHtml.do?orgId=101&tblId=DT_1C86&conn_path=I2)
통계청에서 제공한 2020년 시도별 주민등록 인구현황 데이터를 결합하였습니다.
(출처: https://kosis.kr/statHtml/statHtml.do?orgId=202&tblId=DT_202N_B4&conn_path=I2)

이번 과제를 수행하며 신뢰성 있는 기관으로부터 데이터를 수집하는 과정과 
연구자가 원하는 연구를 수행하기 위해 익힐 간단한 데이터 전처리 과정을 실습했습니다.

os.chdir 함수를 사용한 작업 디렉토리 설정 명령문은 작성자의 개인 디렉토리이므로 
다른 device에서 코드를 실행하시려면 수정이 필요합니다.
"""

### 라이브러리 불러오기
import pandas as pd
import os


### 폴더와 파일명 지정
# 불러올 파일이 위치한 작업 디렉토리 지정
os.chdir('C:/Users/tkdal/OneDrive - 아주대학교/onedrive_pypr/경제정보처리/data_assignment')

# 불러올 파일명 지정
file1_name = '2020_시도별_1인당_지역내총생산__지역총소득__개인소득.xlsx'
file2_name = '2020_시도별_주민등록_인구현황.xlsx'
```

```python
""" 파일 불러오기 """

df1 = pd.read_excel(file1_name)
df2 = pd.read_excel(file2_name)
```

```python
# df1에서 첫 번째 행을 column의 name으로 지정해줍니다.

df1_ad = df1.rename(columns=df1.iloc[0])
df1_ad.head()
```

```python
# 그리고 첫 번째 행을 삭제해줌으로써 데이터만 데이터프레임 내부에 존재하도록 합니다.

df1_com = df1_ad.drop(df1_ad.index[0])
df1_com.head()
```

```python
# df2도 df1과 같은 이슈를 가진 형태이므로 첫 번째 행을 column name으로 지정해줍니다.

df2_ad = df2.rename(columns=df2.iloc[0])
df2_ad.head()
```

```python
# 그리고 데이터프레임 내부에는 데이터만 남겨줍니다.

df2_com = df2_ad.drop(df2_ad.index[0])
df2_com.head()
```

```python
""" 파일을 하나로 merge (df1 + df2 = df) """

df = pd.merge(df1_ad, df2_ad, left_on='시도별', right_on='시·도별(1)')
```

시도별 데이터여서 데이터 수가 주피터 노트북에 표현되지 않을만큼 많지 않기 때문에 전체 데이터를 출력해서 확인해봅니다.

merge의 기준이 되는 시도별 속성을 기준으로 원인을 분석했습니다. 이유를 확인해보니 df1에서 시도별 명칭은 full name으로 작성되었고, df2에서 시도별 명칭은 두 글자까지만 표시된 것을 확인했습니다.

그래서 시도별 명칭을 두 데이터프레임 중 하나의 것으로 통일해줘야겠다고 판단했습니다. 저는 full name으로 표기된 df1의 시도별 명칭으로 통일해주는 전처리를 실행했습니다.

그리고 두 데이터 모두 시도명 정렬 순서가 같기 때문에 별도의 순서 조정은 하지 않아도 되는 것을 확인했습니다.

```python
df1_com.iloc[:,[0]]
```

```python
sido_name = df1_com.iloc[:,[0]]
```

```python
# df2에 df1의 full name으로 된 시도별 명칭이 삽입된 것을 확인할 수 있습니다.

df2_com.iloc[:,[0]] = sido_name
df2_com
```

```python
""" 파일을 하나로 merge (df1 + df2 = df) """

df = pd.merge(df1_com, df2_com, left_on='시도별', right_on='시·도별(1)')
```

```python
df = df.drop(['시·도별(1)'], axis = 'columns')
```
